```python
from flask import Flask, render_template_string, request
from googletrans import Translator, LANGUAGES
import nest_asyncio

# Allow Flask to run in Jupyter Notebook
nest_asyncio.apply()

# Create a Flask app
app = Flask(__name__)

# Initialize the translator
translator = Translator()

# HTML template with CSS for styling
HTML_TEMPLATE = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Language Translator</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            background-color: #4aa1b8;
            margin: 0;
            padding: 0;
            display: flex;
            justify-content: center;
            align-items: center;
            min-height: 100vh;
        }
        .container {
            background: #4ab3b8;
            padding: 20px;
            border-radius: 8px;
            box-shadow: 0 4px 6px rgba(0, 0, 0, 0.1);
            width: 100%;
            max-width: 600px;
            text-align: center;  /* Center content inside the container */
        }
        h1 {
            text-align: center;
            color: #333;
        }
        form {
            margin-top: 20px;
            text-align: center;  /* Center form elements */
        }
        label {
            display: block;
            font-weight: bold;
            margin-bottom: 5px;
        }
        textarea, select {
            width: 80%;  /* Adjust the width of the textarea */
            padding: 10px;
            margin-bottom: 20px;
            border: 1px solid #ddd;
            border-radius: 4px;
            margin-left: auto;
            margin-right: auto;
            display: block;
        }
        button {
            background-color: #4a84b8F;
            color: #104645;
            padding: 10px 20px;
            border: none;
            border-radius: 4px;
            cursor: pointer;
            font-weight: bold;
            margin-left: auto;
            margin-right: auto;
            display: block;
        }
        button:hover {
            background-color: #0056b3;
        }
        .result {
            background: #f9f9f9;
            padding: 10px;
            border: 1px solid #ddd;
            border-radius: 4px;
            margin-top: 20px;
            text-align: left;  /* Align result text to the left */
        }
    </style>
</head>
<body>
    <div class="container">
        <h1>LANGUAGE TRANSLATOR</h1>
        <form method="POST" action="/">
            <label for="text">Enter Text:</label>
            <textarea id="text" name="text" rows="4" required></textarea>
            <label for="language">Select Target Language:</label>
            <select id="language" name="language" required>
                {% for code, name in languages.items() %}
                    <option value="{{ code }}">{{ name }}</option>
                {% endfor %}
            </select>
            <button type="submit">Translate</button>
        </form>
        {% if translated_text %}
            <div class="result">
                <h2>Translated Text:</h2>
                <p>{{ translated_text }}</p>
            </div>
        {% endif %}
    </div>
</body>
</html>
"""

@app.route('/', methods=['GET', 'POST'])
def home():
    translated_text = ""
    if request.method == 'POST':
        input_text = request.form['text']
        target_language = request.form['language']
        try:
            # Perform translation
            translation = translator.translate(input_text, dest=target_language)
            translated_text = translation.text
        except Exception as e:
            translated_text = f"Error: {str(e)}"

    return render_template_string(HTML_TEMPLATE, languages=LANGUAGES, translated_text=translated_text)

# Run the Flask app
if __name__ == '__main__':
    app.run(port=5000)
```
